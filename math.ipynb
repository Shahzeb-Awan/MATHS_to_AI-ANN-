```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import load_model
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_data = pd.read_csv('/content/drive/MyDrive/math/data_Area_Of_circle_train.csv')

# Extract input features and output
X_train = train_data.iloc[:, :-1].values
y_train = train_data.iloc[:, -1].values

# Load testing data
test_data = pd.read_csv('/content/drive/MyDrive/math/data_Area_Of_circle_test.csv')

# Extract input features for testing
X_test = test_data.iloc[:, :-1].values
```

```python
# Normalize the input features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Split the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Build the ANN model
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=3))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Define early stopping criteria
early_stopping = EarlyStopping(patience=10, verbose=1)
```

```python
# Train the model
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=20, callbacks=[early_stopping])

# Save the trained model
model.save('trained_model.h5')

# Load the saved model
model = load_model('trained_model.h5')

# Use the model to make predictions on the testing data
predictions = model.predict(X_test)
```

```python
# Create a DataFrame with the predictions
output_df = pd.DataFrame(predictions, columns=['Predicted Area'])

# Save the predictions to a CSV file
output_df.to_csv('predictions_math.csv', index=False)

!mv predictions_math.csv /content/drive/MyDrive/math/
```
